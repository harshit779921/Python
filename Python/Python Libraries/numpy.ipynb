# Array Manipulation


### Array Manipulation - Changing shape


```python
import numpy as np

a = np.arange(9)
print("The original array:")
print(a)
print()

b = a.reshape(3, 3)
print()
print("Modified Array: ")
print(b)
```

```python
print(b.flatten())
```

```python
b.flatten(order="F")
```

```python
a = np.arange(12).reshape(4, 3)
a
```

```python
print(np.transpose(a))
```

```python
b = np.arange(8).reshape(2, 4)
b
```

```python
c = b.reshape(2, 2, 2)
print(c)
```

```python
np.rollaxis(c, 2, 1)
```

```python
np.swapaxes(c, 2, 1)
```

## Iterating Over Array


```python
a = np.arange(0, 45, 5)
a = a.reshape(3, 3)
a
```

```python
for x in np.nditer(a):
    print(x)
```

### Iteration order (c-style and f-style)


```python
print(a)
for x in np.nditer(a, order="C"):
    print(x)

for x in np.nditer(a, order="f"):
    print(x)
```

### Joining Arrays


```python
a = np.array([[1, 2], [3, 4]])
print("First array:")
print(a)
b = np.array([[5, 6], [7, 8]])
print("Second array:")
print(b)
print("\n")

print(np.concatenate((a, b)))
print("\n")
print(np.concatenate((a, b), axis=1))
```

### Splitting Array


```python
a = np.arange(9)
print(a)

print(np.split(a, 3))
```

```python
print(np.split(a, [4, 5]))
```
